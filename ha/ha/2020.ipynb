```python
import pandas as pd
import numpy as np
```

```python
#강남구
df2020_1_GANGNAM = pd.read_excel('./data/2020/2020_1.xls')

df2020_1_GANGNAM = df2020_1_GANGNAM[['날짜','측정소명','PM10']]
df2020_1_GANGNAM[df2020_1_GANGNAM['측정소명'] == '강남구'] 
df2020_1_GANGNAM['날짜'] = df2020_1_GANGNAM['날짜'].str[:10]
df2020_1_GANGNAM['날짜'] = df2020_1_GANGNAM['날짜'].str.replace('-','')

df2020_1_GANGNAM.drop(df2020_1_GANGNAM.loc[df2020_1_GANGNAM['PM10'] == -999].index, inplace = True)
df2020_1_GANGNAM = df2020_1_GANGNAM.groupby('날짜').mean()
df2020_1_GANGNAM['PM10'] = round(df2020_1_GANGNAM['PM10'], 2)
df2020_1_GANGNAM.rename(columns={'PM10' : '강남구'})
```
