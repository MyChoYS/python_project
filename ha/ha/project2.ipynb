```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./data/2008/2008_1.xls')

df = df[['날짜','측정소명','PM10']]
df[df['측정소명'] == '중구'] 
df['날짜'] = df['날짜'].str[:10]
df['날짜'] = df['날짜'].str.replace('-','')

#결측값(nan이 아니라 -999로 표시됨)
df.drop(df.loc[df['PM10'] == -999].index, inplace = True)
df = df.groupby('날짜').mean()
df['PM10'] = round(df['PM10'], 2)
df.rename(columns={'PM10' : '중구'})
```
